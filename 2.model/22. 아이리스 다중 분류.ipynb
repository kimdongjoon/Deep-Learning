```python
import pandas as pd
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
import pandas as pd
df = pd.read_csv('./dataset/iris.csv', 
                 names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"])
df.head()
```

```python
dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]
Y_obj[0], Y_obj[50], Y_obj[100]
```

```python
Y_obj.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
Y
```

```python
# One hot encoding
Y_encoded = tf.keras.utils.to_categorical(Y)
Y_encoded[0], Y_encoded[50], Y_encoded[100]
```

```python
Y_encoded.shape
```

```python
# seed 값 설정
seed = 0
tf.random.set_seed(seed)
```

```python
# 딥러닝 모델 설정
model = Sequential([
    Dense(16, input_shape=(4,), activation='relu'),
    Dense(3, activation='softmax')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

```python
#X.shape
Y_encoded.shape
```

```python
# 모델 실행
model.fit(X, Y_encoded, epochs=50, batch_size=10)
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model.evaluate(X, Y_encoded, verbose=2)[1]))
```

```python
# 딥러닝 모델 설정
model2 = Sequential([
    Dense(12, input_shape=(4,), activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')
])
model2.summary()
```

```python
# 모델 컴파일 
model2.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 모델 실행
model2.fit(X, Y_encoded, epochs=500, batch_size=30)
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model2.evaluate(X, Y_encoded, verbose=2)[1]))
```
