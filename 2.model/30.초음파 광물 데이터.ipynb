```python
import pandas as pd

df = pd.read_csv('./dataset/sonar.csv', header=None)
df.head()
```

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
dataset = df.values
X = np.array(dataset[:,0:60], dtype=np.float64)
Y_obj = dataset[:,60]
```

```python
type(Y_obj)
```

```python
# 문자열 변환
e = LabelEncoder().fit(Y_obj)
Y = np.array(e.transform(Y_obj), dtype=np.float64)
```

```python
Y_obj[:5], Y[:5]
```

```python
# 모델 설정
model = Sequential([
    Dense(24, input_dim=60, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='binary_crossentropy',   # mean_squared_error
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=5)
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model.evaluate(X, Y, verbose=2)[1]))
```

# 과적합 피하기 - 초음파광물 사례
## 학습셋과 테스트셋으로 구분

```python
import pandas as pd

df = pd.read_csv('dataset/sonar.csv', header=None)
df.head()
```

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
dataset = df.values
X = np.array(dataset[:,0:60], dtype=np.float64)
Y_obj = dataset[:,60]
```

```python
# 문자열 변환
e = LabelEncoder().fit(Y_obj)
Y = np.array(e.transform(Y_obj), dtype=np.float64)
```

```python
# 학습셋과 테스트셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, 
                                                    random_state=seed)
```

```python
# 모델 설정
model = Sequential([
    Dense(24, input_dim=60, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='binary_crossentropy',   # mean_squared_error
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=130, batch_size=5)
```

```python
# 테스트셋에 모델 적용 
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test, verbose=2) [1]))
```

## 모델 저장과 재사용

```python
import pandas as pd

df = pd.read_csv('dataset/sonar.csv', header=None)
df.head()
```

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
dataset = df.values
X = np.array(dataset[:,0:60], dtype=np.float64)
Y_obj = dataset[:,60]
```

```python
# 문자열 변환
e = LabelEncoder().fit(Y_obj)
Y = np.array(e.transform(Y_obj), dtype=np.float64)
```

```python
# 학습셋과 테스트셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, 
                                                    random_state=seed)
```

```python
# 모델 설정
model = Sequential([
    Dense(24, input_dim=60, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='binary_crossentropy',   # mean_squared_error
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 모델을 컴퓨터에 저장
model.save('model/my_model.h5')
```

```python
del model       # 테스트를 위해 메모리 내의 모델을 삭제
```

```python
model = load_model('model/my_model.h5') # 모델을 새로 불러옴
```

```python
# 테스트셋에 불러온 모델 적용 
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test, verbose=2) [1]))
```
