```python
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
# 실행할 때마다 같은 결과를 출력하기 위한 seed 값 설정
seed = 2000
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#준비된 수술 환자 데이터를 불러들임
Data_set = np.loadtxt("./dataset/ThoraricSurgery.csv", delimiter = ",")
Data_set
```

```python
#환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
# X = Data_set[:,0:17]
# Y = Data_set[:,17]
# X.shape, Y.shape

x_data = Data_set[:,0:17]
y_data = Data_set[:,17]
x_data.shape, y_data.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

```

```python
#딥러닝 모델 설정

# model.add(Dense(24, input_shape = (17, ), activation = 'relu'))
# model.add(Dense(8, activation = 'relu'))
# model.add(Dense(1, activation = 'sigmoid'))



model = Sequential ([
    Dense(24, input_shape = (17, ), activation = 'relu'),
    Dense(8, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])
model.summary()
```

```python
# 딥러닝 환경 설정 (오차함수, 최적화 함수)
model.compile(loss = 'binary_crossentropy',
             optimizer = 'adam' , metrics = ['accuracy'])
```

```python
# 딥 러닝 학습
model.fit(x_data,y_data, epochs = 500, batch_size = 30)
```

```python
#결과 출력 
# evaluate() returns the loss value & matrics value for the model
print("\n Accuracy : %.4f" % (model.evaluate(x_data, y_data, verbose = 1)))
```
